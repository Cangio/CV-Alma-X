## Implement a filter for preserving edges
The filter is useful for filtering the image being able to still recognize all the edges.\
Edge recognition is implemented in marker recognition and object recognition. 

```python
# Import libraries once
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_src = cv2.imread("../2-markers/test-imgs/double-rectangle.jpg")

#img_src = cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB)

img_dst = cv2.bilateralFilter(img_src, 9, 75, 75, cv2.BORDER_DEFAULT)

#cv2.imshow('Result Image', img_dst)

plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB))
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_dst, cv2.COLOR_BGR2RGB))
plt.title("Filtered image")
plt.show()

plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img_src[270:400, 80:280, :], cv2.COLOR_BGR2RGB))
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_dst[270:400, 80:280, :], cv2.COLOR_BGR2RGB))
plt.title("Filtered image")
plt.show()

print(img_src.shape)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
