cv.selectROI(	windowName, img[, showCrosshair[, fromCenter]]	)

showCrosshair=True (default)
fromCenter=False

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

image=cv2.imread('../1-box/zimages/marsyard/task-example.jpg')

roi = cv2.selectROI('select single roi',image,False,False)
cv2.destroyAllWindows()

roi=image[roi[1]:roi[1]+roi[3],roi[0]:roi[0]+roi[2]]

plt.imshow(roi)
plt.show()
```

selection of multiple rois:
cv2.selectROIs
cv2.selectROI passing a vector

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

image=cv2.imread('../1-box/zimages/box/box2.jpg')

roi = cv2.selectROIs('select rois',image)
cv2.destroyAllWindows()

for roijj in roi:
    roimg=image[roijj[1]:roijj[1]+roijj[3],roijj[0]:roijj[0]+roijj[2]]
    plt.imshow(roimg)
    plt.show()
```
